```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print(f"First Three Restaurants Sales Data:\n{sales_data[:3]}")
```

```python
# Total sales per year
print("Total Sales per Year:\n",np.sum(sales_data[:,1:],axis=0))
```

```python
# Minimum sales did by restorent
print("Minimum Sales by Restorent:\n",np.min(sales_data[:,1:],axis=1))
```

```python
# Miximum sales did per year
print("Maximum Sales per Year by restorent:\n",np.max(sales_data[:,1:],axis=0))
```

```python
# Average sales per restorent
print("Average Sales per Restorent:\n",np.mean(sales_data[:,1:],axis=1))
```

```python
# Cumulative sales per restorent
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print("Cumulative Sales per Restorent:\n",cumsum)
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cumulative Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Average Cumulative Sales")
plt.grid(True)
plt.show()
```

### Vector Manupulation

```python
v1=np.array([1,2,3,4,5])
v2=np.array([6,7,8,9,10])
```

```python
print("Addition of Vectors:\n",v1+v2)
print("Multiplication of Vectors:\n",v1*v2)
print("Dot Product:",np.dot(v1,v2))
```

```python
# Compute the angle between v1 and v2 in radians
angel=np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))
print("Angle between v1 and v2 in radians:",angel)
```

### Vectorize Method

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper=np.vectorize(str.upper)
print("Vectorized Upper:\n",vectorized_upper(restaurant_types))
```
